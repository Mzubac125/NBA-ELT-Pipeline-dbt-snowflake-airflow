```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date
from bs4 import BeautifulSoup, Comment
```

```python
last_five_years = []
current_year = date.today().year

for i in range(0,5):
    last_five_years.append(int(current_year) - i)
```

```python
URL = f'https://www.basketball-reference.com/contracts/players.html'
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'lxml')
```

```python
results = soup.find('div', id = 'div_player-contracts')
```

```python
table_data = results.find_all('tr')
```

```python
data_rows = []
for row in table_data[2:]:
        #Find all table data
        data = row.find_all('td')
        individual_data = [i.text for i in data]
        #Add data to the empty list
        data_rows.append(individual_data)
```

```python
data_rows
```

```python
#Find all the class names for the headers
data_headers = results.find_all("th", attrs={"data-stat": True})

for th in data_headers:
    print(f"Text: '{th.get_text(strip=True)}' | Class: {th.get('class')} | Data-Stat: {th.get('data-stat')}")
```

```python
#Get a list of all the headers
data_headers = results.find_all('th', class_ = ['center','right', 'over_header center', 'sort_default_asc left'])
for row in data_headers:
    titles = [i.text for i in data_headers]
```

```python
#Get desired headers
titles = titles[33:42]
```

```python
titles
```

```python
df = pd.DataFrame(data_rows, columns= titles)
df.head()
```

```python
df.dtypes
```

```python
def clean_salary(value):
    return float(value.replace('$',"").replace(',',"")) if value else None
```

```python
salary_columns = df.columns[2:]
for col in salary_columns:
    df[col] = df[col].apply(clean_salary)
df.head()
```
